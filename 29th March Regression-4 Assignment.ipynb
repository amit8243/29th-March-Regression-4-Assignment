```python
#Regression-4 Assignment
"""Q1. What is Lasso Regression, and how does it differ from other regression techniques?"""
Ans: Lasso regression is a type of regularized linear regression technique that uses shrinkage. It adds a 
penalty term to the cost function, which is the sum of the squared error terms. This penalty term is the 
sum of the absolute values of the coefficients multiplied by a constant. The effect of this penalty term 
is to shrink some of the coefficients to zero, effectively eliminating them from the model. This is 
different from other regression techniques, such as ridge regression, which use a penalty term that is 
the sum of the squared values of the coefficients. This means that ridge regression does not eliminate 
any coefficients, but instead reduces their magnitude.

"""Q2. What is the main advantage of using Lasso Regression in feature selection?"""
Ans: The main advantage of using Lasso Regression in feature selection is that it can automatically 
perform feature selection by shrinking the coefficients of less important features to zero. This helps to 
reduce the complexity of the model and improve its interpretability. Additionally, it can help to reduce 
overfitting by eliminating irrelevant features.

"""Q3. How do you interpret the coefficients of a Lasso Regression model?"""
Ans: The coefficients of a Lasso Regression model represent the amount of influence each feature has on 
the outcome. A positive coefficient indicates that an increase in the feature will lead to an increase in 
the outcome, while a negative coefficient indicates that an increase in the feature will lead to a 
decrease in the outcome. The magnitude of the coefficient indicates the strength of the relationship 
between the feature and the outcome. The larger the coefficient, the stronger the relationship.

"""Q4. What are the tuning parameters that can be adjusted in Lasso Regression, and how do they affect the
model's performance?"""
Ans: The tuning parameters that can be adjusted in Lasso Regression are the regularization parameter 
(lambda) and the maximum number of iterations. The regularization parameter controls the amount of 
shrinkage applied to the coefficients, and a higher value of lambda will result in more shrinkage. 
The maximum number of iterations determines the number of times the algorithm will run before it stops. 
Increasing the maximum number of iterations can improve the model's performance, but it can also increase 
the time required to train the model.

"""Q5. Can Lasso Regression be used for non-linear regression problems? If yes, how?"""
Ans: Yes, Lasso Regression can be used for non-linear regression problems. It can be used by transforming 
the non-linear features into linear ones using polynomial features or other transformations. This allows 
the Lasso Regression to fit a non-linear model to the data.

"""Q6. What is the difference between Ridge Regression and Lasso Regression?"""
Ans: Ridge Regression is a type of regularized linear regression that adds a penalty term to the cost 
function to reduce the complexity of the model and prevent overfitting. It shrinks the coefficients of 
correlated features towards each other, but does not set any coefficient to zero.

Lasso Regression is also a type of regularized linear regression that adds a penalty term to the cost 
function to reduce the complexity of the model and prevent overfitting. It shrinks the coefficients of 
correlated features towards each other and can set some coefficients to zero, effectively removing them 
from the model.

"""Q7. Can Lasso Regression handle multicollinearity in the input features? If yes, how?"""
Ans: Yes, Lasso Regression can handle multicollinearity in the input features by shrinking the 
coefficients of correlated variables to zero. This is done by adding a penalty term to the cost 
function which penalizes large coefficients. This forces some of the correlated variables to have a 
coefficient of zero, thus eliminating the multicollinearity.

"""Q8. How do you choose the optimal value of the regularization parameter (lambda) in Lasso Regression?"""
Ans:The optimal value of the regularization parameter (lambda) in Lasso Regression can be chosen using 
cross-validation. Cross-validation is a technique used to evaluate predictive models by partitioning the
original sample into a training set to train the model, and a test set to evaluate it. The model is then 
evaluated by computing the error rate on the test set. The optimal value of lambda can be chosen by 
finding the value of lambda that minimizes the error rate on the test set.
```
